```python
# Imports
import pandas as pd
from datetime import datetime
import os
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Préparer le fichier CSV ---
if not os.path.exists("pointage.csv"):
    df = pd.DataFrame(columns=["Nom", "Code", "Date", "Heure", "Type"])
    df.to_csv("pointage.csv", index=False)

# --- Liste des employés ---
employes = {
    "A123": "Alice",
    "B456": "Bob",
    "C789": "Charlie"
}

# --- Mot de passe manager ---
MOT_DE_PASSE_MANAGER = "admin123"  # change selon ton choix

# --- Fonction de pointage ---
def pointer(code_personnel, type_pointage):
    if code_personnel not in employes:
        return "Code invalide !"
    nom = employes[code_personnel]
    df = pd.read_csv("pointage.csv")
    nouvelle_ligne = {
        "Nom": nom,
        "Code": code_personnel,
        "Date": datetime.now().strftime("%Y-%m-%d"),
        "Heure": datetime.now().strftime("%H:%M:%S"),
        "Type": type_pointage
    }
    df = pd.concat([df, pd.DataFrame([nouvelle_ligne])], ignore_index=True)
    df.to_csv("pointage.csv", index=False)
    return f"{nom} a pointé {type_pointage} à {nouvelle_ligne['Heure']}"

# --- Interface employé ---
code_input = widgets.Text(description="Code :")
type_input = widgets.Dropdown(options=["entrée", "sortie"], description="Type :")
button_pointage = widgets.Button(description="Pointer")
output_employe = widgets.Output()
historique_output = widgets.Output()

def afficher_historique(code_personnel):
    df = pd.read_csv("pointage.csv")
    df = df[df["Code"] == code_personnel]
    with historique_output:
        clear_output()
        if df.empty:
            print("Aucun pointage enregistré.")
        else:
            print("Derniers pointages :")
            print(df.tail(5))

def on_button_pointage(b):
    with output_employe:
        clear_output()
        message = pointer(code_input.value, type_input.value)
        print(message)
        afficher_historique(code_input.value)

button_pointage.on_click(on_button_pointage)

employe_box = widgets.VBox([code_input, type_input, button_pointage, output_employe, historique_output])

# --- Interface manager ---
manager_password_input = widgets.Password(description="Mot de passe :")
button_manager = widgets.Button(description="Afficher toutes les données")
output_manager = widgets.Output()

def on_button_manager(b):
    with output_manager:
        clear_output()
        if manager_password_input.value != MOT_DE_PASSE_MANAGER:
            print("Mot de passe manager incorrect !")
            return
        df = pd.read_csv("pointage.csv")
        if df.empty:
            print("Aucun pointage enregistré.")
        else:
            print(df)

button_manager.on_click(on_button_manager)
manager_box = widgets.VBox([manager_password_input, button_manager, output_manager])

# --- Affichage final ---
display(widgets.Tab([employe_box, manager_box]))
display(widgets.HTML("<b>Onglets :</b> 0=Employés, 1=Manager"))
```
