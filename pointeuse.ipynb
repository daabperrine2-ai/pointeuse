```python
# Imports
import pandas as pd
from datetime import datetime
import os
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Étape 1 : Préparer le fichier CSV ---
if not os.path.exists("pointage.csv"):
    df = pd.DataFrame(columns=["Nom", "ID", "Date", "Heure", "Type"])
    df.to_csv("pointage.csv", index=False)

# --- Étape 2 : Fonction de pointage ---
def pointer(nom, id_personne, type_pointage):
    df = pd.read_csv("pointage.csv")
    nouvelle_ligne = {
        "Nom": nom,
        "ID": id_personne,
        "Date": datetime.now().strftime("%Y-%m-%d"),
        "Heure": datetime.now().strftime("%H:%M:%S"),
        "Type": type_pointage
    }
    df = pd.concat([df, pd.DataFrame([nouvelle_ligne])], ignore_index=True)
    df.to_csv("pointage.csv", index=False)
    return f"{nom} a pointé {type_pointage} à {nouvelle_ligne['Heure']}"

# --- Étape 3 : Créer l’interface ipywidgets ---
nom_input = widgets.Text(description="Nom :")
id_input = widgets.Text(description="ID :")
type_input = widgets.Dropdown(options=["entrée", "sortie"], description="Type :")
button = widgets.Button(description="Pointer")
output = widgets.Output()

def on_button_clicked(b):
    with output:
        clear_output()
        message = pointer(nom_input.value, id_input.value, type_input.value)
        print(message)

button.on_click(on_button_clicked)

display(widgets.VBox([nom_input, id_input, type_input, button, output]))
```
